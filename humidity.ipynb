```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

tem = 15

def e_T(tem):
    return 6.1078 * 10 ** (7.5*tem / (tem + 237.3))
```

```python
e_T(tem)
```

```python
def a_T(tem, e_T):
    return 217 * e_T / ( tem + 273.15)
```

```python
a_T(tem,e_T(tem))
```

```python
## plotting 
tem = -50
p_list = []
tem_list = []
for i in range(100):
    p_list.append(a_T(tem,e_T(tem)))
    tem += 1
    tem_list.append(tem)
```

```python
plt.xlabel("temperature")
plt.ylabel("a(T)[g / m^3]")
plt.title("absolute humidty")
plt.plot(tem_list,p_list)
plt.show()
```

```python
print(p_list[-1])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(tem_list,p_list, 'go--')
ax.set_title("dashed absolute humidity")
ax.set_xlabel("temperature")
ax.set_ylabel("a(T)[g / m^3]")
plt.show()
```

```python
## いくつ水分子があるか計算
g_mol = 18.01528  # g / mol, 1 mol あたりのgram数
Na = 6.02214076 * 10 ** 23  #1 molあたりの個数
item_total = p_list[-1] / g_mol * Na # mol数を数えて、それから個数を計算
print("temperature is ",tem_list[-1])
print("gram is ",p_list[-1])
print(item_total)
```
