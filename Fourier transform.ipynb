```python
import matplotlib.pyplot as plt
import cmath
import math
%matplotlib inline
```

```python
def function(x: int) -> int:
    return cmath.sin(x)

#def fourier(function, x:float,  t: int) -> float:
def fourier(x:int,  t: int):
    return cmath.exp(-2 * complex(0,1) * t)
```

```python
ans = []
for x,t in zip(range(100),range(150)):
    ans.append(function(math.radians(x)) * fourier(x, t)) 
```

```python
ans[32]
```

```python
rel, img = ans[0].real, ans[0].imag
```

```python
rels, imgs = [], []
for pairs in ans:
    rel, img = pairs.real, pairs.imag
    rels.append(rel)
    imgs.append(img)
```

```python
plt.plot(rels,imgs,animated=True)
```
